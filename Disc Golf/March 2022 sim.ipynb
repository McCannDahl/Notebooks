### This is a website used to simulate real world disc golf throws. It uses complex ODE calculations and a series of inital conditions to predict the arodynamic trajectory of vairious discs. Simply adjust the sliders to see how the disc will fly.

```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

button = widgets.Button(description='Refresh')
display(button)

fig, ax = plt.subplots(2,1)

def get_fig():
    ax[0].plot(np.arange(10), np.random.random(10))
    ax[1].plot(np.arange(10), np.random.random(10),color='red')

get_fig()

def on_button_clicked(b):
    ax[0].cla()
    ax[1].cla()
    get_fig()
    
button.on_click(on_button_clicked)
```
