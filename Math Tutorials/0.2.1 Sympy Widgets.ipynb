# Sympy Widgets
##### There are a lot of widgets to try here https://ipywidgets.readthedocs.io/en/latest/examples/Widget%20List.html

Widgets take the form of graphical outputs & inputs. They could be a slider that sets a variable or a plot of a function. These are really handy when taking inputs from non-technical audiences and showing analysis in a clear summarized way.

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from sympy import *
from sympy.plotting import *
from sympy.matrices import *
init_printing()
import ipywidgets as widgets
```

```python
x = symbols("x")
```

Move the slider and reexecute the plot to see the changes

```python
x_slider = widgets.FloatSlider(
    value=7.5,
    min=0,
    max=10.0,
    step=0.1,
    description='y:'
)
display(x_slider)
```

```python
p = plot(x*x_slider.value, (x, -5, 5))
```
