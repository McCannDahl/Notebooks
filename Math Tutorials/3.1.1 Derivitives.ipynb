## Definition
Derivatives are super handy because they show us how something is changing with resepct to other things.

## Example
For example, if I was a cop and I knew where a car was going to be on a highway, I could use derivitives to calculate his speed and acceleration because....

```python
acceleration, velocity, position = symbols('acceleration velocity position')
display(Eq(velocity,Derivative(position,t)))
display(Eq(acceleration,Derivative(velocity,t)))
```

Therefore,

```python
display(Eq(acceleration,Derivative(position,t,2)))
```

If the car's position looked like this...

```python
x = t**3/4
display(Eq(position,x,evaluate=False))
p = plot(x, (t, 0, 10), xlabel='Time (s)', ylabel='Position (m)')
```

Then to find the velociy of the car (or how fast it is going at a particular time), all we need to do is take the derivitive with respect to t

```python
v = Derivative(x, t)
display(Eq(velocity,v))
v = v.doit()
display(Eq(velocity,v))
p = plot(v, (t, 0, 10), xlabel='Time (s)', ylabel='Velocity (m)')
```

Likewise, we can find the acceleration of the car at any given time if we derive the velocity

```python
a = Derivative(v, t)
display(Eq(acceleration,a))
a = a.doit()
display(Eq(acceleration,a))
p = plot(a, (t, 0, 10), xlabel='Time (s)', ylabel='Acceleration (m)')
```

So it looks like the acceleration is increasing at a constant rate, meaning the person driving is moving his foot on the pedal at a constant rate. On the velocy graph, you can see about after 6 seconds he breaks the speed limit of 70 mph (30m/s)
