## Definition
Integrals are super handy because they are a way to infinitely sum something to a discrete value.
They can be seen as "reversing" a derivitive.
They can also be seen as an "area under the curve".

## Example
For example, if I was a cop and I knew how much a car was accelerating, I could use integrals to calculate his speed and position because....

```python
acceleration, velocity, position = symbols('acceleration velocity position')
display(Eq(velocity,Integral(acceleration,t)))
display(Eq(position,Integral(velocity,t)))
```

Therefore,

```python
display(Eq(position,Integral(acceleration,t,t)))
```

If the car's position looked like this...

```python
a = 3*t/2
display(Eq(acceleration,a,evaluate=False))
p = plot(a, (t, 0, 10), xlabel='Time (s)', ylabel='Acceleration (m)')
```

Then to find the velociy of the car (or how fast it is going at a particular time), all we need to do is take the integral with respect to t

```python
v = Integral(a, t)
display(Eq(velocity,v))
v = v.doit()
display(Eq(velocity,v))
p = plot(v, (t, 0, 10), xlabel='Time (s)', ylabel='Velocity (m)')
```

Likewise, we can find the position of the car at any given time if we integrate the velocity

```python
x = Integral(v, t)
display(Eq(position,x))
x = x.doit()
display(Eq(position,x))
p = plot(x, (t, 0, 10), xlabel='Time (s)', ylabel='Position (m)')
```

So it looks like the acceleration is increasing at a constant rate, meaning the person driving is moving his foot on the pedal at a constant rate. On the velocy graph, you can see about after 6 seconds he breaks the speed limit of 70 mph (30m/s), and after 10 seconds he's traveled 250m.
